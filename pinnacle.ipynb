```python
'''                               PREMATCH pinnacle NAVIGATOR''' 


url_pin_prematch = 'https://www.pinnacle.com/en/soccer/matchups/highlights'
#changing chromedriver default options
options = Options()
options.add_argument('window-size=1920x1080') #Headless = True
options.headless = True #for not opening the navigator
#Headless = True

pin_prematch = webdriver.Chrome(r'C:\Users\KNL\Desktop\Ale\Upwork\Scraping_bets\chromedriver.exe', options=options)
pin_prematch.get(url_pin_prematch)
```

```python
''' SCRAPE AND LOAD INTO df_pre_pinnacle ''' 
def pinnacle_prematch():
    i=2
    df_pre_pinnacle=pd.DataFrame(columns=['teams', '1_pin','X_pin','2_pin', 'goals',  'over','under'])
    while i<30:
        i+=1
        try:
            #home_team = pin_prematch.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/main/div/div[2]/div[{}]'.format(i)).text.split('\n')[0]
            #away_team = pin_prematch.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/main/div/div[2]/div[{}]'.format(i)).text.split('\n')[1]
            teams = pin_prematch.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/main/div/div[2]/div[{}]'.format(i)).text.split('\n')[0:2]
            h_wins = pin_prematch.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/main/div/div[2]/div[{}]'.format(i)).text.split('\n')[3]
            X      = pin_prematch.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/main/div/div[2]/div[{}]'.format(i)).text.split('\n')[4]
            a_wins = pin_prematch.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/main/div/div[2]/div[{}]'.format(i)).text.split('\n')[5]
            
            goals = pin_prematch.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/main/div/div[2]/div[{}]'.format(i)).text.split('\n')[10]
            over  = pin_prematch.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/main/div/div[2]/div[{}]'.format(i)).text.split('\n')[11]
            under = pin_prematch.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/main/div/div[2]/div[{}]'.format(i)).text.split('\n')[12]
            
            df_pre_pinnacle.loc[len(df_pre_pinnacle.index)] = [teams,  h_wins, X, a_wins, goals,  over, under]

            
        except:
            pass

    return df_pre_pinnacle
    pinnacle_prematch()
```

```python
''' TENIS NAVIGATOR '''
options = Options()
options.headless = True #for not opening the navigator
#options.add_argument('window-size=1920x1080') #Headless = True
url_pin_tenis = 'https://www.pinnacle.com/en/tennis/matchups/live/'
pin = webdriver.Chrome(r'C:\Users\KNL\Desktop\Ale\Upwork\Scraping_bets\chromedriver.exe', options=options)

pin.get(url_pin_tenis)
pin.set_window_size(1920, 1080)
```

```python
''' tenis live function'''
def pinnacle_tenis():
    df_pin_tenis = pd.DataFrame(columns=['players_pin', '1_pin', '2_pin', 'games_pin','over_pin', 'under_pin'])
    for i in range(50):
        try:
            elemento = pin.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/main/div/div[2]/div[{}]'.format(i)).text.split('\n')
            #print(elemento)
            if len(elemento)>=6:
                if 'Sets' in elemento[0]:
                    players = elemento[0].replace('(Sets)','')+' vs '+elemento[1].replace('(Sets)','')
                    pin_1 = elemento[-3]
                    pin_2 = elemento[-2]
                    #print(players, '   ||  ', elemento[-3],'|', elemento[-2])
                    #print('/*********', i)
                if 'Games' in elemento[0]:
                    #print(elemento[-4], '|', elemento[-2],'|', elemento[-3])
                    games_pin = elemento[-3]
                    over_pin = elemento[-4]
                    under_pin = elemento[-2]
                    #print('-------------------')
                df_pin_tenis.loc[len(df_pin_tenis.index)] = [players, pin_1, pin_2, games_pin, over_pin, under_pin]
        except:
            pass
    return df_pin_tenis.drop_duplicates('players_pin')
```
