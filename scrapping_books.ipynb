## Funciones de scraper

```python
def get_urls(url):
    root_url = 'https://books.toscrape.com/' 
    links_categorias = '//ul[@class="nav nav-list"]/li//ul/li/a/@href'

    r = requests.get(url)
    home = r.content.decode('utf-8')
    parser = html.fromstring(home)
    
    categorias_url=parser.xpath(links_categorias)
    categorias_url = [root_url+x for x in categorias_url]
    return categorias_url
```

```python
def parser_content(urls):
    cont_dic = {}
    r = requests.get(urls)
    titulo =  '//article[@class="product_pod"]//h3/a/text()'
    precio = '//div[@class="product_price"]/p[@class="price_color"]/text()'
    #print(req.status_code)
    # Es el valor que estamos buscando cuando estamos haciendo una peticion a una página web
    home = req.content.decode('utf-8')
    parser = html.fromstring(home)
    if r.status_code ==200:
        titulos_book=parser.xpath(titulo)
        precios = parser.xpath(precio)
        cont_dic['Title'] = titulos_book
        cont_dic['Price'] = precios
    return cont_dic
```

```python
links_entregables = get_urls(url)
links_entregables[:3]
```

```python
len(links_entregables)
```

```python
data = []
for idx, i in enumerate(links_entregables):
    print('Se está scrapeando la pag: ', idx)
    data.append(parser_content(i))
```

### Funcion para obtener todos los links de libros de una sopa

```python
# funcion de obtener links
def url_items(sopa, url):
    url_root = 'https://books.toscrape.com/' 
    lista_articulos = s.find_all('article', class_='product_pod')
    links_libros = [x.find('h3').find('a').get('href') for x in lista_articulos]
    links_libros = [urljoin(url_root, i) for i in links_libros]
    return links_libros
```

#### Genero una sopa para cada pagina, aplico la funcion para obtener los links pelados de cada contenedor en cada pagina y cambiar a la pagina siguiente.
#### Esto retorna una lista con cada lista de links

```python
links_items = []
it = 0
while it<50:
    it+=1
    print(f'Estoy en la pagina {url_inicial}')
    r_pag=requests.get(url_inicial)
    s_p=BeautifulSoup(r_pag.text, 'lxml')
    links= url_items(s_p, r_pag)
    next_a = s_p.select('li.next > a') # Esto es para ir consiguiendo la siguiente pagina con el metodo select
    if not next_a or not next_a[0].get('href'):
        break
    url_inicial = urljoin(url_inicial, next_a[0].get('href'))
    links_items.append(links)
```

```python
def scraper_book(url):
    content_book = {}
    r = requests.get(url)
    s_item = BeautifulSoup(r.text,'lxml')

    titulo = s_item.find('h1').get_text(strip=True) #Me elimina los espacios del final y el inicio
    if titulo:
        content_book['Titulo'] = titulo
    else:
        content_book['Titulo'] = None
    
    precio = s_item.find('p', class_='price_color').get_text()
    if precio:
        content_book['Precio'] = precio
    else:
        content_book['Precio'] = 0
    
    
    ancla_desc = s_item.find('div' , id="product_description") # voy a usar un metodo para anclarme a la etiqueta vecina de la decripcion porque no tiene 'class=..' , entonces uso la de al lado tambien
    descripcion = ancla_desc.find_next_sibling('p').get_text()
    if descripcion:
        content_book['Descripcion'] = descripcion
    else:
        content_book['Descripcion'] = None

    src_img = s_item.find('div', class_='item active').find('img').get('src')
    src_img = urljoin(url_root, src_img)
    if descripcion:
        content_book['Imagen'] = src_img
    else:
        content_book['Imagen'] = None

    return content_book    
```

```python
list_scrapper = list_scrapper[:10]
datos_book = []
for idx, i in enumerate(list_scrapper):
    print(f'estas escrapeando la pag {idx}')
    datos_book.append(scraper_book(i))
```
