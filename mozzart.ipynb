```python
'''Navigator Options               MOZZART LIVE'''
options = Options()
options.headless = True #for not opening the navigator
options.add_argument('window-size=1920x1080') #Headless = True
url_mozart = 'https://www.mozzartbet.com/en#/live/sport/1'
moz = webdriver.Chrome(r'C:\Users\KNL\Desktop\Ale\Upwork\Scraping_bets\chromedriver.exe', options=options)
moz.get(url_mozart)
```

```python
'''                                     MOZZART LIVE FUNCTION '''
def mozzart():
    #options = Options()
    #options.headless = True #for not opening the navigator
    #options.add_argument('window-size=1920x1080') #Headless = True
    #url_mozart = 'https://www.mozzartbet.com/en#/live/sport/1'
    #moz = webdriver.Chrome(r'C:\Users\KNL\Desktop\Ale\Upwork\Scraping_bets\chromedriver.exe', options=options)
    #moz.get(url_mozart)
    i=0
    df_mozart = pd.DataFrame(columns=['teams', 'final_result', 'over_under']) 
    while i<30:
        
        i=i+1
        try:
            partido = moz.find_element(By.XPATH, '/html/body/div[1]/div[1]/section/section/div[1]/section/section/div/div[1]/div/section/div/article[{}]'.format(i)).text.split('\n')
            if len(partido)==29:
                teams=partido[2]+'  '+partido[3]
                final_result= partido[10:13]
                over_under=partido[19:22]
                df_mozart.loc[len(df_mozart.index)] = [teams, final_result, over_under] 
                
                
                #print(teams, final_result, over_under)
                
            if len(partido)==27:
                teams=partido[2]+'  '+partido[3]
                final_result= partido[8:11]
                over_under=partido[17:20]
                df_mozart.loc[len(df_mozart.index)] = [teams, final_result, over_under] 
                
                #print(teams, final_result, over_under)

        except:
            break
    
    
    #moz.quit()
    
    '''
    TRANSFORMING AND REPLACING DATA
    '''
    # Spliting the numbers

    df_mozart[['1_moz', 'X_moz', '2_moz']] = df_mozart.final_result.apply(lambda x: pd.Series(str(x).split(',')))
    df_mozart[['goals_mz', 'under_mz','over_mz']] = df_mozart.over_under.apply(lambda x: pd.Series(str(x).split(',')))

    #Deleting some characters

    cols = ['1_moz', 'X_moz', '2_moz','teams','goals_mz', 'under_mz','over_mz']
    for col in cols:
        df_mozart[col]=df_mozart[col].apply(lambda x: x.replace("'",""))
        df_mozart[col]=df_mozart[col].apply(lambda x: x.replace("[",""))
        df_mozart[col]=df_mozart[col].apply(lambda x: x.replace("]",""))
        df_mozart[col]=df_mozart[col].apply(lambda x: x.replace("-","1"))
        
    col2 = cols = ['1_moz', 'X_moz', '2_moz','goals_mz', 'under_mz','over_mz']
    for col in col2:    
        df_mozart[col]= df_mozart[col].astype('float64')

    df_mozart.drop(columns=['final_result', 'over_under'], inplace=True)

    
    return df_mozart

```
