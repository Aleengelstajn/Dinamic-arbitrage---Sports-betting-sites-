```python
'''                                  PINNCALE NAVIGATOR TO THE BACKGROUND'''


url_pin = 'https://www.pinnacle.com/en/soccer/matchups/live'

#changing chromedriver default options
options = Options()
options.headless = True #for not opening the navigator
options.add_argument('window-size=1920x1080') #Headless = True
pin = webdriver.Chrome(r'C:\Users\KNL\Desktop\Ale\Upwork\Scraping_bets\chromedriver.exe', options=options)
pin.get(url_pin)
```

```python
'''           function LIVE PINNACLE'''


def pinnacle():
    #pin = webdriver.Chrome(r'C:\Users\KNL\Desktop\Ale\Upwork\Scraping_bets\chromedriver.exe', options=options)
    #pin.get(url_pin)
    j = 1 # ITS 2(AND IN THE BEGGINING OF THE LOOP IT WILL BECOME 3) BECAUSE BEFORE ARE OTHER ELEMENTS OF THE PAGE THAT WE DON'T NEED
    df_pinnacle=pd.DataFrame(columns=['teams_pin', '1_pin', 'X_pin', '2_pin','goals_pin', 'over_pin', 'under_pin'])
    B=True
    while j<45:
        j+=1
        try:
            partido = pin.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/main/div/div[2]/div[{}]'.format(j)).text.split('\n')
            teams = partido[0]+' '+partido[2]

            home_wins= partido[5]
            X = partido[6]
            away_wins = partido[7]

            goals = partido[12]
            under= partido[13]
            over = partido[15]
            if len(partido)==18:
                teams = partido[0]+' '+partido[3]

                home_wins= partido[6]
                X = partido[7]
                away_wins = partido[8]

                goals = partido[13]
                under= partido[14]
                over = partido[16]
            
            df_pinnacle.loc[len(df_pinnacle.index)] = [teams, home_wins, X, away_wins, goals,under, over]
            #print(teams, home_wins, X, away_wins, goals, over, under)
        except:
            pass
    #pin.quit()
    cols = ['1_pin', 'X_pin', '2_pin','goals_pin', 'over_pin', 'under_pin']
    
    for col in cols:
        for j in range(len(df_pinnacle.index)):
            if len(df_pinnacle[col].loc[j])>6:
                df_pinnacle[col].loc[j] = 1
    for col in cols:
            df_pinnacle[col]=df_pinnacle[col].astype('float64')
    
    return df_pinnacle
```

```python
''' SOCCERBET FUNCTION'''
def soccerbet():
    df_socc = pd.DataFrame(columns=['teams', '1_socc', 'X_socc', '2_socc', 'goals_socc', 'under_socc', 'over_socc'])
    for i in range(60):
        try:
            element = soc.find_element(By.XPATH, f'//*[@id="result-tables"]/div[2]/div[1]/table/tbody/tr[{i}]').text.split('\n')
            teams = element[1][4:].replace(':','')
            teams = re.sub('\d+','',teams)

            socc_1 = element[2].replace('-', '1')
            socc_X = element[3].replace('-', '1')
            socc_2 = element[4].replace('-', '1')

            goals_socc= element[10][:3].replace('-', '1')
            under_socc= element[8].replace('-', '1')
            over_socc=element[9].replace('-', '1')

            df_socc.loc[len(df_socc.index)] = [teams, socc_1, socc_X,socc_2, goals_socc, under_socc, over_socc]

            #print(teams, socc_1, socc_X, socc_2, '|',under_socc, over_socc)

            cols = ['1_socc', 'X_socc', '2_socc', 'goals_socc', 'under_socc',
       'over_socc']
            for c in cols:
                df_socc[c] = df_socc[c].astype('float64')
        except:
            pass
    return df_socc
```

```python
def admiral():
    
    df_adm = pd.DataFrame(columns=['teams', '1_adm', 'X_adm', '2_adm', 'goals_adm', 'under_adm', 'over_adm'])
    for k in range(40):
        for j in range(6):
            try:
                elemento = adm.find_element(By.XPATH, f'/html/body/app-root/app-betting-live/div[2]/section[2]/div/app-sport-groups-container/app-sport-group[1]/section/app-competition-group[{k}]/section/div[2]/app-event-live[{j}]').text.split('\n')
                
                teams = elemento[2]+' '+ elemento[3]

                adm_1 = elemento[8].replace('-','1')
                adm_X = elemento[9].replace('-','1')
                adm_2 = elemento[10].replace('-','1')
                
                goals_adm = elemento[20].replace('-','0')
                
                under = elemento[19].replace('-','1')
                over = elemento[21].replace('-','1')
                df_adm.loc[len(df_adm.index)] = [teams, adm_1, adm_X, adm_2, goals_adm, under, over]
                #print(teams,'|', adm_1, adm_X, adm_2,'    |  ', goals_adm, under ,over)
            except:
                pass
    cols = ['1_adm', 'X_adm', '2_adm',  'under_adm','goals_adm','over_adm']
    for c in cols:
        df_adm[c]= df_adm[c].astype('float64')

    df_adm = df_adm[df_adm.goals_adm<5]
    return df_adm
```
