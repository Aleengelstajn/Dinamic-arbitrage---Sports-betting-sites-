```python
'''           function LIVE PINNACLE'''


def pinnacle():
    #pin = webdriver.Chrome(r'C:\Users\KNL\Desktop\Ale\Upwork\Scraping_bets\chromedriver.exe', options=options)
    #pin.get(url_pin)
    j = 1 # ITS 2(AND IN THE BEGGINING OF THE LOOP IT WILL BECOME 3) BECAUSE BEFORE ARE OTHER ELEMENTS OF THE PAGE THAT WE DON'T NEED
    df_pinnacle=pd.DataFrame(columns=['teams_pin', '1_pin', 'X_pin', '2_pin','goals_pin', 'over_pin', 'under_pin'])
    B=True
    while j<45:
        j+=1
        try:
            partido = pin.find_element(By.XPATH, '/html/body/div[2]/div/div[2]/main/div/div[2]/div[{}]'.format(j)).text.split('\n')
            teams = partido[0]+' '+partido[2]

            home_wins= partido[5]
            X = partido[6]
            away_wins = partido[7]

            goals = partido[12]
            under= partido[13]
            over = partido[15]
            if len(partido)==18:
                teams = partido[0]+' '+partido[3]

                home_wins= partido[6]
                X = partido[7]
                away_wins = partido[8]

                goals = partido[13]
                under= partido[14]
                over = partido[16]
            
            df_pinnacle.loc[len(df_pinnacle.index)] = [teams, home_wins, X, away_wins, goals,under, over]
            #print(teams, home_wins, X, away_wins, goals, over, under)
        except:
            pass
    #pin.quit()
    cols = ['1_pin', 'X_pin', '2_pin','goals_pin', 'over_pin', 'under_pin']
    
    for col in cols:
        for j in range(len(df_pinnacle.index)):
            if len(df_pinnacle[col].loc[j])>6:
                df_pinnacle[col].loc[j] = 1
    for col in cols:
            df_pinnacle[col]=df_pinnacle[col].astype('float64')
    
    return df_pinnacle
```

```python
'''Navigator Options               MOZZART LIVE'''
options = Options()
options.headless = True #for not opening the navigator
options.add_argument('window-size=1920x1080') #Headless = True
url_mozart = 'https://www.mozzartbet.com/en#/live/sport/1'
moz = webdriver.Chrome(r'C:\Users\KNL\Desktop\Ale\Upwork\Scraping_bets\chromedriver.exe', options=options)
moz.get(url_mozart)
```

```python
'''                                     MOZZART LIVE FUNCTION '''
def mozzart():
    #options = Options()
    #options.headless = True #for not opening the navigator
    #options.add_argument('window-size=1920x1080') #Headless = True
    #url_mozart = 'https://www.mozzartbet.com/en#/live/sport/1'
    #moz = webdriver.Chrome(r'C:\Users\KNL\Desktop\Ale\Upwork\Scraping_bets\chromedriver.exe', options=options)
    #moz.get(url_mozart)
    i=0
    df_mozart = pd.DataFrame(columns=['teams', 'final_result', 'over_under']) 
    while i<30:
        
        i=i+1
        try:
            partido = moz.find_element(By.XPATH, '/html/body/div[1]/div[1]/section/section/div[1]/section/section/div/div[1]/div/section/div/article[{}]'.format(i)).text.split('\n')
            if len(partido)==29:
                teams=partido[2]+'  '+partido[3]
                final_result= partido[10:13]
                over_under=partido[19:22]
                df_mozart.loc[len(df_mozart.index)] = [teams, final_result, over_under] 
                
                
                #print(teams, final_result, over_under)
                
            if len(partido)==27:
                teams=partido[2]+'  '+partido[3]
                final_result= partido[8:11]
                over_under=partido[17:20]
                df_mozart.loc[len(df_mozart.index)] = [teams, final_result, over_under] 
                
                #print(teams, final_result, over_under)

        except:
            break
    
    
    #moz.quit()
    
    '''
    TRANSFORMING AND REPLACING DATA
    '''
    # Spliting the numbers

    df_mozart[['1_moz', 'X_moz', '2_moz']] = df_mozart.final_result.apply(lambda x: pd.Series(str(x).split(',')))
    df_mozart[['goals_mz', 'under_mz','over_mz']] = df_mozart.over_under.apply(lambda x: pd.Series(str(x).split(',')))

    #Deleting some characters

    cols = ['1_moz', 'X_moz', '2_moz','teams','goals_mz', 'under_mz','over_mz']
    for col in cols:
        df_mozart[col]=df_mozart[col].apply(lambda x: x.replace("'",""))
        df_mozart[col]=df_mozart[col].apply(lambda x: x.replace("[",""))
        df_mozart[col]=df_mozart[col].apply(lambda x: x.replace("]",""))
        df_mozart[col]=df_mozart[col].apply(lambda x: x.replace("-","1"))
        
    col2 = cols = ['1_moz', 'X_moz', '2_moz','goals_mz', 'under_mz','over_mz']
    for col in col2:    
        df_mozart[col]= df_mozart[col].astype('float64')

    df_mozart.drop(columns=['final_result', 'over_under'], inplace=True)

    
    return df_mozart
```
